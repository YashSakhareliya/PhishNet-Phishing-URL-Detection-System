## Check for Missing Value

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
print(df['Label'].value_counts())
sns.countplot(data=df, x='Label')
plt.show()
```

```python
sns.histplot(df['Total-character'], kde=True)
plt.show()
```

```python
corr_metrix = df[['Total-character', 'Uppercase_letter', 'Lowercase_letter',
       'Total_numbers', 'Total_symbol', 'Total_slesh', 'Total_dots',
       'Total_desh']].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_metrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# relation ship between feature
sns.scatterplot(data=df, x='Total-character', y='Total_symbol', hue='Label')
plt.show()
```

```python
sns.scatterplot(data=df, x='Uppercase_letter', y='Total_symbol', hue='Label')
plt.show()
```
